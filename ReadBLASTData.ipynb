```python
import pygetdata as gd
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u
from astropy import constants as c
```

```python
blast2012path = 'data/blast2012/'
```

```python
data = gd.dirfile(blast2012path+'blast_2012.z', gd.RDONLY)
```

```python
t_chin = data.getdata('T_CHIN')
```

```python
t1_prime = data.getdata('T_1_PRIME')
t2_prime = data.getdata('T_2_PRIME')
t1_sec = data.getdata('T_1_SECOND')
t2_sec = data.getdata('T_2_SECOND')
```

```python
t = data.getdata('TIME')
```

```python
dday = ((t - t[0])*u.s).to(u.day)
```

```python
plt.figure(figsize=(15,5))
plt.plot(dday.to(u.hr), t1_prime, label='T_1_PRIME')
plt.plot(dday.to(u.hr), t2_prime, label='T_2_PRIME')
plt.plot(dday.to(u.hr), t1_sec, label='T_1_SECOND')
plt.plot(dday.to(u.hr), t2_sec, label='T_2_SECOND')
for d in np.arange(14):
    plt.axvline(d*24, linestyle='--', color='gray')
plt.legend()
plt.ylim([-40,20])
plt.ylabel('Hours since launch')
plt.ylabel(r'Temperature ($^o$C)')
plt.savefig('primary_and_secondary_temperatures.pdf')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(dday.to(u.hr), t1_prime - t2_prime, label=r'$\Delta T$ Primary')
plt.plot(dday.to(u.hr), t1_sec - t2_sec, label=r'$\Delta T$ Secondary')
for d in np.arange(14):
    plt.axvline(d*24, linestyle='--', color='gray')
plt.legend()
plt.ylim([-2,2])
plt.ylabel('Hours since launch')
plt.ylabel(r'Temperature ($^o$C)')
plt.savefig('primary_and_secondary_gradients.pdf')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(dday.to(u.hr), t1_prime - t1_sec, label=r'$\Delta T$ Primary - Secondary')
for d in np.arange(14):
    plt.axvline(d*24, linestyle='--', color='gray')
plt.legend()
plt.ylim([-5,20])
plt.xlabel('Hours since launch')
plt.ylabel(r'Temperature ($^o$C)')
plt.savefig('primary_to_secondary_gradient.pdf')
plt.show()
```
